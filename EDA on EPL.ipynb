# **Column Definition**

- `Season`: Season year(s)
- `Pos`: final position that season
- `Team`: name of team
- `Pld`: Number of matches played
- `W`: Number of wins
- `D`: Number of draws
- `L`: Number of losses
- `GF`: Goals scored that season
- `GA`: Goals conceded that season
- `GD`: Difference in goals scored vs. conceded
- `Pts`: Total points at end of season
- `Qualification or Relegation`: result at end of season

**Write a function to rename results in the qualification column to aid legibility**

```python
epl_data['Qualification or relegation'].unique()
```

```python
def result (qualify_status):
  if 'Champions League' in qualify_status:
    return 'Champions League'
  elif 'Europa' in qualify_status or 'UEFA' in qualify_status:
    return 'Europa League'
  elif 'Relegation' in qualify_status:
    return 'Relegated'
  elif 'Intertoto' in qualify_status:
    return 'Intertoto'
  else:
    return '-'
  return qualify_status
```

```python
epl_data['Qualification or relegation'] = epl_data['Qualification or relegation'].apply(result)
```

```python
epl_data['Qualification or relegation'].value_counts()
```

*The code below filters the EPL dataset to include only teams that qualified for the Champions League and then calculates and prints aggregated statistics*

```python
champions_league = epl_data[epl_data['Qualification or relegation'] == 'Champions League']
champions_league_stats = champions_league \
                          .groupby('Season') \
                          .agg({'Pos':'max', 'Pts':'min', 'GD':'min'})


print(champions_league_stats)


```

**Group by Team and Aggregate**



```python

Grouping_stats = epl_data \
                          .groupby('Team') \
                          .agg({'Pos':'min', 'Pts':'sum', 'GF': 'sum'})

print(Grouping_stats.sort_values('Pts', ascending=False).reset_index())
```

*The code below filters the dataset to isolate the league winners (teams finishing in 1st place) and the teams that were relegated (teams finishing in 18th place) in each season. It then plots the points of these teams over different seasons, providing a visual comparison of the performance (in terms of points) of the league winners and the relegated teams. This allows for an easy comparison of the points gap between the top and bottom teams in the league across various seasons.*

```python
epl_winner = epl_data[epl_data['Pos'] == 1]
epl_winner = epl_winner.reset_index(drop=True)


epl_relegation = epl_data[epl_data['Pos'] == 18]
epl_relegation = epl_relegation.reset_index(drop=True)

plt.figure(figsize=(18, 9))
sns.lineplot(x='Season', y='Pts', data=epl_winner, label='Winner')

sns.lineplot(x='Season', y='Pts', data=epl_relegation, label='Relegation')
plt.xticks(rotation=90)
plt.title('Winner vs Relegation')
plt.xlabel('Season')
plt.ylabel('Points')
```

**Examine how GD and Pts are linked**



```python
chelsea = epl_data[epl_data['Team'] == 'Chelsea']
chelsea


plt.figure(figsize=(18, 9))
ax=sns.lineplot(x='Season', y='GD', data=chelsea, label = 'Goal Diff', marker='o')
ax2 = ax.twinx()
sns.lineplot(x='Season', y='Pts', data=chelsea, label='Points', ax=ax2, marker='o', color='green')

lines, labels = ax.get_legend_handles_labels()
lines2, labels2 = ax2.get_legend_handles_labels()

ax.legend(lines + lines2, labels + labels2, loc='upper left')

ax2.get_legend().remove()

ax.set_xticklabels(ax.get_xticklabels(), rotation=90)

plt.title("Chelsea GD and Pts per season")
plt.show()

```
